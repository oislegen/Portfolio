````python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
````

## Exploration of Toronto Neighborhoods

Let's create a function to repeat the same process of extracting the nearby venues in each neighborhood to all the neighborhoods with the word "Toronto" in its borough name.

````python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
````

Use the above function for each neighborhood and store the data about nearby venues in a data frame. 

````python
toronto_venues = getNearbyVenues(names=toronto_data['Neighborhood'],
                                   latitudes=toronto_data['Latitude'],
                                   longitudes=toronto_data['Longitude']
                                  )
````

Let's find out how many unique categories can be curated from all the returned venues.

One hot encode ```Venue Category``` feature and construct a data frame with the neighborhood of each venue in the first column and the venue category in the other columns. Note that there is a venue category "Neighborhood". Therefore, we will call the venue category ```Neighborhood Category``` not to confuse it with the first column ```Neighborhood```.

````python
# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

#Change the name of the "Neighborhood" column in venue categories to "Neighborhood Category" 
toronto_onehot.rename({'Neighborhood': 'Neighborhood Category'}, axis='columns', inplace=True)

#Insert a new columm "Neighborhood" into the first column of the dataframe
toronto_onehot.insert(0, 'Neighborhood', toronto_venues['Neighborhood'] )

toronto_onehot.head()
````

Check the size of the above data frame.

````python
toronto_onehot.shape
````

Group rows by neighborhood and take the mean frequency of occurrence for each venue category.

````python
toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()
toronto_grouped
````

Check the size of the above data frame.

````python
toronto_grouped.shape
````

Print each neighborhood along with the ten most common venue category in the neighborhood.

````python
num_top_venues = 10

for hood in toronto_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
````

Write a function to sort the venues in descending order. If there are fewer venues than ```num_top_venues```, those venues will be called "None". 

````python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    for ind in np.arange(0,row_categories_sorted.shape[0]):
        if row_categories_sorted[ind]==0:
            row_categories_sorted.index.values[ind]="None"
    return row_categories_sorted.index.values[0:num_top_venues]
````

Create the new data frame and display the top ten venues for each neighborhood. Note that the data frame will show "None" in the relevant cells if there are fewer venues than ten in a neighborhood.

````python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted
````

## Cluster Neighborhoods

Run the k-means clustering algorithm to divide the neighborhoods into 5 clusters according to the frequency of occurrence of the venue categories in each neighborhood.

````python
# set number of clusters
kclusters = 5

toronto_grouped_clustering = toronto_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:20] 
````

Create a new data frame that includes the cluster as well as the top venue category for each neighborhood.

````python
# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

toronto_merged = toronto_data

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

toronto_merged.head() 
````

Visualize the resulting clusters.

````python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighborhood'], toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
````

## Conclusion

As we checked above with the number of venues in each neighborhood in the ```toronto_venues``` dataframe, the first cluster contains neighborhoods which contain more than four nearby venues except "The Beaches". The Beaches neighborhood resembles the other neighborhoods in the first cluster as it has a nearby coffee shop and a pub. The clusters other than the first cluster contain neighborhoods with 4 or fewer nearby venues with no coffee shops or pubs nearby which suggests they are more residential neighborhoods, some of them including venues related to children.   

Kmeans clustering is very sensitive to outliers so the neighborhoods with too few venues are separated from the first cluster mainly containing the neighborhoods with many nearby venues.
