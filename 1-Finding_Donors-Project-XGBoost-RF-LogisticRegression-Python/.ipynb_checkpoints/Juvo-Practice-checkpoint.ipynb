```python
import sklearn 

import pandas as pd

import numpy as np

import matplotlib 
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
sns.set()

%matplotlib inline

import pandas_profiling
```

```python
def to_binary(string):
    if string=='>50K':
        return 1
    else: 
        return 0
```

```python
raw_income=raw_data['income']
income_binary=raw_income.apply(to_binary)
income_binary
```

```python
raw_features=raw_data.drop('income',axis=1)
raw_features
```

```python
numeric_columns=raw_features.select_dtypes(include=['int','float']).columns.values.tolist()
f,ax=plt.subplots(figsize=(6,6))
ax=sns.heatmap(raw_features[numeric_columns].corr(), cmap='viridis_r',annot=True)
```

```python
f, ax=plt.subplots(len(numeric_columns),2,figsize=(9,len(numeric_columns)*6))
for i, label in enumerate(numeric_columns):
    sns.distplot(raw_features[label], kde=False, vertical=True, hist_kws={'alpha':1},ax=ax[i,0])
    sns.boxplot(y=raw_features[label], x=income_binary, ax=ax[i,1])
    ax[i,0].set_xlabel('Count')  
f.tight_layout()    
f.show()    
```

```python
sns.pairplot(raw_features[numeric_columns])
```

```python
# Logtransform
raw_features[['capital-gain','capital-loss']]=raw_features[['capital-gain','capital-loss']].apply(lambda x: np.log(x+1))
```

```python
# Scale the numeric columns
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
raw_features[numeric_columns]=scaler.fit_transform(raw_features[numeric_columns])
```

```python
categorical_columns=raw_features.select_dtypes(include=['object']).columns.values.tolist()
```

```python
base_color=sns.color_palette()[0]
f,ax=plt.subplots(len(categorical_columns),2, figsize=(9,len(categorical_columns)*6))
for i,label in enumerate(categorical_columns):
    order=raw_features[label].value_counts().index
    sns.countplot(data=raw_features, y=label, order=order, color=base_color, ax=ax[i][0])
    sns.barplot(x=income_binary, y=raw_features[label],order=order,color=base_color,ax=ax[i][1])
f.tight_layout()
f.show()
```

```python
raw_features['race'][['race'].str.contains('Amer-Indian-Eskimo|Other')]='Other'
```

```python
features_final=pd.get_dummies(raw_features)
feature_names=features_final.columns.values.tolist()
feature_names
```

```python
from sklearn.model_selection import train_test_split

X_pretrain, X_test,y_pretrain, y_test=train_test_split(features_final,income_binary, test_size=0.2, random_state=0)
X_train, X_valid, y_train, y_valid=train_test_split(X_pretrain, y_pretrain, test_size=0.2, random_state=0)
```

```python
import xgboost as xgb
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

clf_A=xgb.XGBClassifier(random_state=0)
clf_B=RandomForestClassifier(random_state=0)
clf_C=LogisticRegression(random_state=0)
```

```python
from sklearn.metrics import fbeta_score, accuracy_score
results={}
for clf in [clf_A,clf_B,clf_C]:
    clf.fit(X_train,y_train)
    train_predict=clf.predict(X_train)
    valid_predict=clf.predict(X_valid)
    results[clf.__class__.__name__]=['F-score and accuracy for training data:', fbeta_score(y_train,train_predict,beta=0.5),accuracy_score(y_train,train_predict),'F-score and accuracy for validation data:', fbeta_score(y_valid,valid_predict,beta=0.5),accuracy_score(y_valid,valid_predict)]
results    
```

```python
#sns.barplot
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer 

score=make_scorer(fbeta_score, beta=0.5)

clf=xgb.XGBClassifier(random_state=0)

parameters={'max_depth':[1,2], 'learning_rate':[0.03,0.1] ,'n_estimators':[200,300]}

grid_obj=GridSearchCV(clf, parameters,scoring=score, cv=3)
grid_fit=grid_obj.fit(X_train,y_train)

best_clf=grid_fit.best_estimator_

valid_predict=best_clf.predict(X_valid)
print(fbeta_score(y_valid,valid_predict,beta=0.5))
```

```python
from sklearn.metrics import confusion_matrix

test_predict=best_clf.predict(X_test)
print(fbeta_score(y_test,test_predict,beta=0.5))
confusion_matrix(y_test,test_predict)
```

```python
# from sklearn.utils.multiclass import unique_labels
def plot_confusion_matrix(y_true, y_pred, classes,
                          normalize=False,
                          title=None,
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if not title:
        if normalize:
            title = 'Normalized confusion matrix'
        else:
            title = 'Confusion matrix, without normalization'

    # Compute confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    # Only use the labels that appear in the data
    classes = classes[unique_labels(y_true, y_pred)]
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    fig, ax = plt.subplots()
    im = ax.imshow(cm, interpolation='nearest', cmap=cmap)
    ax.figure.colorbar(im, ax=ax)
    # We want to show all ticks...
    ax.set(xticks=np.arange(cm.shape[1]),
           yticks=np.arange(cm.shape[0]),
           # ... and label them with the respective list entries
           xticklabels=classes, yticklabels=classes,
           title=title,
           ylabel='True label',
           xlabel='Predicted label')

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
             rotation_mode="anchor")

    # Loop over data dimensions and create text annotations.
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, format(cm[i, j], fmt),
                    ha="center", va="center",
                    color="white" if cm[i, j] > thresh else "black")
    fig.tight_layout()
    return ax


np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plot_confusion_matrix(y_test,test_predict, classes=raw_income,
                      title='Confusion matrix, without normalization')

# Plot normalized confusion matrix
plot_confusion_matrix(y_test, test_predict, classes=raw_income, normalize=True,
                      title='Normalized confusion matrix')

plt.show()
```
