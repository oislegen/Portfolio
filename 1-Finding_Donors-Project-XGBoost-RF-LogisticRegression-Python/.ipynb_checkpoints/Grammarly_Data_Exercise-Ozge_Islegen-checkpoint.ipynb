```python
# Import package for the decomposition plot
from statsmodels.tsa.seasonal import seasonal_decompose  
def analyze_feature(series,string):
    """
    Draw a barplot for a series
    """
    # Define the color and font size for the plot below
    base_color = sns.color_palette()[0]
   
    f= plt.figure(figsize=(8,6))
    sns.barplot(x=pd.Series(series.index).dt.day,y=series.values, color=base_color)
    plt.ylabel('Number of {} Users'.format(string))
    plt.xlabel('Day of the Month in February 2016')
    f.show() 
    
    # Draw a decomposition plot
    decomposition = seasonal_decompose(series, model='additive')
    fig = decomposition.plot()
    rcParams['figure.figsize'] = 16, 8
    plt.show()
    return                
```

## 2. Calculate the daily retention curve for users who used the app for the first time on the following dates: Feb 4th, Feb 10th, and Feb 14th. Also, show the number of users from each cohort. Daily retention curve is defined as the % of users from the cohort, who used the product that day.

I assume that the retention curve describes the classical retention rate on day N, the proportion of users who come back to the app on Day+N, where Day is the day the users first started using the app.

I define a function below, daily_retention, which first creates a user cohort by choosing the unique users who started using the app on the cohort_day, then iterates through the following days until the end of the month to find the daily retention rate, the percentage of users in the user cohort who used the app i days after the cohort_day (not necessarily the first time):

```python
def daily_retention(cohort_day,days_in_month=29):
    """
    Define a function which calculates the daily retention until the end of the month 
    cohort_day: The day of the month the app was used for the first time by the users
    days_in_month: The number of days in the month 
    """
    # Find the users who used the app for the first time on cohort_day
    user_cohort=feb_data[(feb_data.index.day==cohort_day) & (feb_data['isFirst']==True)]['uid']
    cohort_size=user_cohort.shape[0]
    retention={}
    for i in range(1,days_in_month-cohort_day+1):
        # Day i retention
        retention[i]= feb_data[(feb_data.index.day==cohort_day+i) & feb_data['uid'].isin(set(user_cohort))]['uid'].nunique()/cohort_size *100   
    return cohort_size, retention
```

The function below plots the percent retention give the cohort day and retention dictionary defined in the daily_retention function:

```python
def retention_plot(cohort_day, retention):
    """
    Define a function which plots the % retention given the cohort day and retention dictionary
    """
    # Define the color and font size for the plot below
    base_color=sns.color_palette()[0]

    f=plt.figure()
    plt.plot(list(retention.keys()) ,list(retention.values()), color=base_color, marker='o', linestyle='dashed', \
             linewidth=2, markersize=11)
    plt.ylabel('% Daily Retention')
    plt.xlabel('Day of Retention')
    plt.title('Percent Daily Retention for the February {}th Cohort'.format(cohort_day))
    plt.xticks(np.arange(len(retention)+1), np.arange(len(retention)+1))
    plt.xlim(0.5,len(retention)+0.5)
    f.show()
```

### February 4th Cohort:

```python
# Return cohort size and retention curve 
feb4th_cohort_size, feb4th_retention=daily_retention(4)
print('February 4th cohort size:', feb4th_cohort_size)
print('The daily retention for the remaining days of the month:', feb4th_retention)
```

```python
# Plot the retention curve
retention_plot(4,feb4th_retention)
```

February 4th cohort size is 2735. It looks like the percent retention drops to 55% on the first day and gradually decreases in a weekly cyclic manner afterwards. After the first week, the pattern seems to have settled. During a weekly cycle, the lowest proportion of the cohort use the app on Saturdays (e.g., the 9th day after the first-time ping on the graph or the 13th of February) and the highest proportion of cohort use the app on Monday, Tuesday, Wednesday and Thursdays. This is consistent with the drop we see in the number of active users on Saturdays in Question 1. The lowest percent retention we see is about 30% on the 23th day after the first-time ping.

### Average Number of Pings  Per Day Per User

As the final metric, I calculate the average number of pings per day per user given their traffic source. 
I chose this metric because 1) I do not want it to depend on the number of users who come from a traffic source. I already show the number of users from each traffic source above and I want to compare all traffic sources on a level playing field. 2) I do not want the metric to be biased against new users: 
More than one third of the users are new users who pinged for the first time on a day during February. If a user first pinged on February 28th, I will calculate this user's number of pings on February 28th and 29th and divide it by two days rather than the whole month.

```python
# Create a new column with the index of the dataframe utmsource_data
utmsource_data['ping_date']=utmsource_data.index
utmsource_data.head()
```

```python
# Find the total number of pings per user along with their traffic source, 
# if they are a new user and their first active date
pings_active_days_per_user=utmsource_data[['uid','utmSource','isFirst','ping_date']].groupby('uid').agg({'uid':'count','utmSource':'first','isFirst':'sum','ping_date':'first'})

# Show the first first rows of the resulting dataframe
pings_active_days_per_user.head()
```

```python
# Create a new column which holds the number of active days of a user
# Assign all active days to 29, the number of days in February
pings_active_days_per_user['active_days']=29

# If the user is new (pinged for the first time in February), change the active days to 29 minus the day of the month the user was first active plus 1.
pings_active_days_per_user['active_days'][pings_active_days_per_user['isFirst']==1] \
 =29-pings_active_days_per_user['ping_date'].dt.day+1

pings_active_days_per_user.tail()
```

```python
# Create a new column with average pings per active day for each user
pings_active_days_per_user['avg_daily_pings']=pings_active_days_per_user['uid']/pings_active_days_per_user['active_days']
pings_active_days_per_user.tail()
```

```python
# Find the mean number of daily pings per user coming from a traffic source
average_daily_pings_per_user=pings_active_days_per_user.groupby('utmSource')[['avg_daily_pings']].mean()
average_daily_pings_per_user=average_daily_pings_per_user.sort_values(by='avg_daily_pings',ascending=False)
```

Similarly, blog, bing, Sarah+, twitter show promise in terms of drawing users who will have a high average number of daily pings so it would be a good idea to find way to attract more users to these channels. Gsp, card and display seem to show not much potential in terms of drawing users with a high level of presence so there is no need to extend our efforts in this direction.
